```python
# Import necessary libraries for displaying images
from PIL import Image
import matplotlib.pyplot as plt
from ultralytics import RTDETR
from kaggle_secrets import UserSecretsClient
import wandb
```

```python
user_secrets = UserSecretsClient()
my_secret = user_secrets.get_secret("wandb")
wandb.login(key=my_secret)
# Build a YOLOv9c model from scratch
model = RTDETR("rtdetr-l.pt")

# Display model information (optional)
model.info()

# Train the model on the COCO8 example dataset for 100 epochs
results = model.train(data="/kaggle/input/valorant-enemy-final-dataset/data.yaml", epochs=100, imgsz=640)

# Run inference with the YOLOv9c model on an image
results = model("/kaggle/input/test-image/test_image.png")
```
